```python
from sympy import symbols, expand, factor, diff, sin, exp, integrate, limit, solve, Function
from sympy import dsolve, Eq, Matrix, oo, simplify
x, y = symbols('x y')
expr = x + 2*y
expanded_expr = expand(x*expr)
expanded_expr, factor(expanded_expr)

# derivate 
expr_2 = sin(x) * exp(x) 
d_expr_2 = diff(sin(x)*exp(x), x)

# integral 
i_expr_2 = integrate(d_expr_2)
i_expr_2

# definite integral 
i_inf_expr = integrate(sin(x**2), (x, -oo, oo))
i_inf_expr

# limits 
lim_expr = limit(sin(x)/x, x, 0)
lim_expr

# solve quadratic equaitons 
quad_roots = solve(x**2 - 2, x)
quad_roots

# solve differential equation 
y = Function('y')
t = symbols('t')
diff_soln = dsolve(Eq( y(t).diff(t,t) - y(t), exp(t) ), y(t))
diff_soln

# find the eigenvalues  
two_mat = Matrix([[1,2], [2,2]])
eigvals = two_mat.eigenvals()
eigvals

# gotchas 

# substitute values 
expr_3 = x + 5 
answr = expr_3.subs(x, 2)
answr

# equality relation 
expr_4 = Eq(x + 44, 5)
expr_4

# simplify 
a = (x + 1)**2 
b = x**2 + 2*x + 1
simplify(a - b)
```
